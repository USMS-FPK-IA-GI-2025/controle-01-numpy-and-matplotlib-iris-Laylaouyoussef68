# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
#importation du Numpy et Matplotlib
import numpy as np
import matplotlib as plt
```

```python
# chargez le fichier csv
data = np.genfromtxt('/content/Iris.csv',delimiter=',',skip_header=1)
print(data.shape)
```

L'output **(150, 6)** signifie que l'array data contient **150 lignes** (soit les 150 échantillons de fleurs d'Iris) et **6 colonnes** (les 4 mesures de sépale/pétale, la colonne de l'espèce variety, plus une colonne d'index/ID)

### A2. Séparation des features et des espèces
Séparez :
- les 4 premières colonnes numériques dans un array `features`
- la dernière colonne (espèces) dans un array `species`
Utilisez le slicing.


```python
# Separez les 4 premieres colonnes dans features
features = data[:,1:5]
#array species contiet la derniere colonne
species = data[:,-1]
#afficher taille de features
print("shape de features : ",features.shape)
print("\n")
#afficher taille de species
print("shape de species : ",species.shape)
```

* Le résultat de **features.shape** est **(150, 4)**, ce qui confirme que l'opération de slicing a correctement extrait les 4 colonnes

* L'affichage de ***species.shape*** est ***(150,)***, ce qui indique qu'un vecteur unidimensionnel de 150 éléments a été créé en extrayant la dernière colonne.


### A3. Indexing
Affichez les 5 premières lignes de `features` en utilisant l’indexing.


```python
# l'affichage des 5 premieres lignes de features
print(features[0:5,:])
```

Chacune des 5 lignes affichées représente un échantillon de fleur d'Iris, et les **quatre** **valeurs** sur chaque ligne correspondent aux caractéristiques numériques suivantes :

Cet affichage confirme que le slicing effectué à l'étape A2 a été correct, et que les données numériques sont correctement structurées dans l'array features


### A4. Extraction d’une colonne
Créez un array `sepal_length` contenant uniquement la première colonne de `features`.

```python
# creation d'array sepal_length
sepal_length = features[:,0:1]
#affichage des 5 premiere ligne de sepal_length
print(sepal_length[0:5,])
#afficher size de sepal_length
print("size de sepal_length : ",sepal_length.size)
print("type de sepal_length : ",sepal_length.dtype)
```

# Partie B — Statistiques Basiques et Opérations Mathématiques

### B6. Moyenne par colonne
Calculez la moyenne de chaque colonne de `features` avec `np.mean(axis=0)`.


```python
#la moyenne de chaque colonne de features
moy =  np.mean(features,axis=0)
print(moy)
```


Ce vecteur de 4 valeurs représente les caractéristiques moyennes de l'ensemble du dataset Iris :


*   La première valeur (5.84333333 cm) est la **longueur moyenne du sépale**
*   La deuxième valeur (3.054   cm) est **la largeur moyenne du sépale**


*  La troisième valeur ( 3.75866667 cm) est la **longueur moyenne du pétale.**
*  La quatrième valeur (1.19866667 cm) est la **largeur moyenne du pétale.**





### B7. Min et Max
Trouvez la valeur minimale et maximale de la longueur des pétales.


```python
#afficher la valeur minimale de la longueur des petales
min_pet = np.min(features[:,2:3])
print("la valeur minimale de la longueur des petales : ",min_pet)
#afficher la valeur maximale de la longueur des petales
max_pet = np.max(features[:,2:3])
print("la valeur maximale de la longueur des petales est : ",max_pet)
```

La **valeur minimale** de **1.0** et la **valeur maximale** de **6.9** centimètres définissent l'intervalle dans lequel se situent toutes les observations de longueurs de pétales

### B9. Normalisation
Normalisez les longueurs des pétales en les divisant par leur maximum.
Stockez le résultat dans `normalized_petals`.


```python
#normalisation
normalized_petals = features[:,2:3]/max_pet
print("longuer des petales normaliser est : \n",normalized_petals)
```

L'opération a consisté à normaliser les longueurs de pétales en divisant chaque valeur par la valeur maximale trouvée précédemment (max_pet = 6.9). Le résultat stocké dans normalized_petals est un nouvel array où toutes les valeurs sont désormais comprises entre 0 et 1

### B10. Somme cumulée
Calculez la somme cumulée des largeurs des pétales pour Versicolor.
Affichez les 10 premières valeurs.


```python
#la somme cummulee
#stocker les largeurs des petales dans ver_pet_wi
ver_pet_wi = features[50:100,3:4]
print("size de ver_pet_wi : ",ver_pet_wi.shape)
sum_cumm = np.cumsum(ver_pet_wi)
print("la somme cumuee est : ",sum_cumm)
```

### B10. Somme cumulée
Calculez la somme cumulée des largeurs des pétales pour Versicolor.
Affichez les 10 premières valeurs.


```python
#la somme cummulee
#stocker les largeurs des petales dans ver_pet_wi
ver_pet_wi = features[50:100,3:4]
print("size de ver_pet_wi : ",ver_pet_wi.shape)
sum_cumm = np.cumsum(ver_pet_wi)
print("la somme cumuee est : ",sum_cumm)
```

L'opération np.cumsum() sur la largeur des pétales de l'espèce Versicolor (lignes 50 à 99) a généré un vecteur de 50 éléments où chaque valeur représente la somme progressive des largeurs de pétales observées jusqu'à cet échantillon. Le résultat final, 66.3, est la somme totale des largeurs de pétales pour les 50 fleurs Versicolor.
