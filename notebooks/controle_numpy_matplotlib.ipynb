### E24. Bar chart avec error bars
Moyennes par espèce avec écart-type.


```python
species_names = np.unique(species)
means = []
stds = []
#calculer moyennes et ecarte-types par specie
for sp in species_names:
    petal_lengths = features[species == sp, 2]
    means.append(np.mean(petal_lengths))
    stds.append(np.std(petal_lengths))
#bar chart avec error bars
plt.figure()
plt.bar(species_names,means,yerr=stds)
plt.ylabel("Longueur des petals")
plt.xlabel("Especes")
plt.title("moyenne et ecart-type de la longueur des petales par espece")
plt.show()
```

Le graphique en barres représente la longueur moyenne des pétales pour chaque espèce, tandis que les barres d’erreur indiquent l’écart-type, c’est-à-dire la dispersion des valeurs autour de la moyenne.

On observe que l’espèce Iris-setosa possède les pétales les plus courts et une faible variabilité, alors que Iris-versicolor et Iris-virginica présentent des longueurs moyennes plus élevées, avec une dispersion plus importante.

### F27. Corrélations
Calculez la matrice de corrélation avec `np.corrcoef`.


```python
corr_matrix = np.corrcoef(features, rowvar=False)

print(corr_matrix)
```

La matrice de corrélation met en évidence les relations entre les différentes caractéristiques.

On observe notamment **une forte corrélation** positive entre la longueur et la largeur des pétales, ce qui signifie que ces deux variables évoluent ensemble.
Les caractéristiques liées aux sépales présentent des corrélations plus modérées

### F28. Visualisation des corrélations
Visualisez les corrélations sous forme de bar chart ou scatter.


```python
labels = [
    "Long. sépales",
    "Larg. sépales",
    "Long. pétales",
    "Larg. pétales"
]

# Corrélation avec la longueur des pétales (colonne 2)
corr_with_petal_length = corr_matrix[2]

plt.figure()
plt.bar(labels, corr_with_petal_length)
plt.ylabel("Coefficient de corrélation")
plt.title("Corrélation avec la longueur des pétales")
plt.show()
```

Le graphique montre que **la longueur des pétales** est fortement corrélée avec **la largeur des pétales**, tandis que les corrélations avec les caractéristiques des sépales sont plus faibles.

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
variances = []
#variance setosa
var_setosa = np.var(features[0:50],axis=0).mean()
variances.append(var_setosa)
#variance versicolor
var_versicolor = np.var(features[50:100],axis=0).mean()
variances.append(var_versicolor)
#variance virinica
var_vurginica = np.var(features[100:],axis=0).mean()
variances.append(var_vurginica)
#specie avec la plus grande variance
max_var_species = unique_species[np.argmax(variances)]
print(max_var_species)
```

```python
#visualisation
plt.figure(figsize=(10, 4))

for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.hist(features[species == max_var_species, i], bins=10)
    plt.title(f"Caractéristique {i+1}")

plt.suptitle(f"Distributions pour {max_var_species}")
plt.tight_layout()
plt.show()
```

L’espèce présentant la plus grande variance globale est **Iris-virginica**, ce qui indique une plus grande dispersion des valeurs pour ses caractéristiques.

Les histogrammes montrent une étendue plus large des distributions, traduisant une variabilité morphologique plus importante.
